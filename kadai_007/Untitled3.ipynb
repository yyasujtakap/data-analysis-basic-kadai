## カテゴリーの出現頻度をカウントし可視化

### Subtask:
`カテゴリー` 列のユニークな出現頻度をカウントし、その結果を棒グラフで表示します。


**Reasoning**:
Count the occurrences of each category in the 'カテゴリー' column and then create a bar plot to visualize these frequencies.



```python
category_counts = df['カテゴリー'].value_counts()

plt.figure(figsize=(10, 6))
category_counts.plot(kind='bar')
plt.title('カテゴリー別出現頻度')
plt.xlabel('カテゴリー')
plt.ylabel('出現頻度')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

## 商品番号ごとの基本統計量算出

### Subtask:
商品番号ごとにグループ化し、単価、在庫、注文数などの数値列の基本統計量 (平均、中央値、標準偏差など) を算出します。


**Reasoning**:
グループ化したい列と集計したい数値列を指定し、基本統計量を算出します。



```python
grouped_stats = df.groupby('商品番号')[['単価', '在庫', '注文数']].agg(['mean', 'median', 'std', 'sum', 'count', 'min', 'max'])
display(grouped_stats)
```
